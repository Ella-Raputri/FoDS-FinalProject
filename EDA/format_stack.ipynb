```python
import pandas as pd
```

```python
df = pd.read_csv('yearly_stack_overflow_count.csv')
df
```

```python
df_melt = df.melt(id_vars="programming language", var_name='year', value_name='count')
df_melt["year"] = df_melt["year"].str.extract("(\d+)$").astype(int)

df_pivot = df_melt.pivot(index="year", columns="programming language", values="count")
df_pivot = df_pivot.loc[:, (df_pivot != 0).any(axis=0)]
df_pivot
```

```python
df_pivot.to_csv('format_stack.csv')
```
