```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('result3.csv')
df
```

```python
# All Tiobe Index Ratings

plt.figure(figsize = (6,4))
df[["tiobe index ratings"]].boxplot()
plt.title("Boxplot for Each Programming Language's \nTiobe Index Ratings 2024", fontsize = 14)
plt.xlabel("Variable", fontsize = 12)
plt.ylabel("Amount", fontsize = 12)
plt.show()
```

```python
# All NonZero Tiobe Index Ratings

plt.figure(figsize = (6,4))
nonzero = df[df["tiobe index ratings"] > 0]
nonzero[["tiobe index ratings"]].boxplot()

plt.title("Boxplot for Each Programming Language's \nTiobe Index Ratings 2024 (Nonzero)", fontsize = 14)
plt.xlabel("Variable", fontsize = 12)
plt.ylabel("Percentage", fontsize = 12)
plt.show()
```
