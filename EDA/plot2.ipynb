```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import ast
```

```python
df = pd.read_csv('job_result2.csv')
```

```python
job_type_counts = df.groupby('programming language')[['manager','full stack','front end','back end','software developer/engineer','mobile developer','devops','game developer','QA engineer','data scientist','data analyst','machine learning engineer','AI engineer','programmer','other analyst','other developer','other engineer','others']].sum()
```

```python
job_type_counts['total'] = job_type_counts.sum(axis=1)
sorted_job_types = job_type_counts.sort_values('total', ascending=False)

split_size = len(sorted_job_types) // 5
first_part = sorted_job_types.iloc[:split_size]
second_part = sorted_job_types.iloc[split_size:2*split_size]
third_part = sorted_job_types.iloc[2*split_size:3*split_size]
fourth_part = sorted_job_types.iloc[3*split_size:4*split_size]
fifth_part = sorted_job_types.iloc[4*split_size:]
```

```python
def plot_job_types(part, title):
    plt.figure(figsize=(15, 8))  
    
    colors = sns.color_palette("tab20", n_colors=part.shape[1] - 1) 
    part.drop('total', axis=1).plot(kind='bar', stacked=True, width=0.8, ax=plt.gca(), color=colors)
    
    plt.ylabel('Programming Language')
    plt.xlabel('Count of Job Types') 
    plt.title(title) 
    plt.legend(title='Job Type', bbox_to_anchor=(1.05, 1), loc='upper left')  
    plt.tight_layout()
    plt.show()
```

```python
plot_job_types(first_part, 'Job Type Count by Programming Language (Part 1)')
```

```python
plot_job_types(second_part, 'Job Type Count by Programming Language (Part 2)')
```

```python
plot_job_types(third_part, 'Job Type Count by Programming Language (Part 3)')
```

```python
plot_job_types(fourth_part, 'Job Type Count by Programming Language (Part 4)')
```

```python
plot_job_types(fifth_part, 'Job Type Count by Programming Language (Part 5)')
```
