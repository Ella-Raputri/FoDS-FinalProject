```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('result3.csv')
df
```

```python
filtered_df = df[(df.select_dtypes(include=['float64', 'int64']) > 0).all(axis=1)]

numeric_columns = filtered_df.select_dtypes(include=['float64', 'int64']).columns

plt.figure(figsize=(15, 10))
for i, column in enumerate(numeric_columns, 1):
    plt.subplot(3, 3, i)
    sns.histplot(filtered_df[column], bins=10, kde=True, color='skyblue', edgecolor='black')

    plt.title(f"Histogram with Trend Line of {column}")
    plt.xlabel(column)
    plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
for column in numeric_columns:
    plt.figure(figsize=(6, 4))
    sns.histplot(filtered_df[column], bins=10, kde=True, color='skyblue', edgecolor='black')
    plt.title(f"Histogram with Trend Line of {column}")
    plt.xlabel(column)
    plt.ylabel("Frequency")
    plt.show()
```
