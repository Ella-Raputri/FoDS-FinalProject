```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
# tiobe index
tiobe_df = pd.read_csv('tiobeindex2.csv')
```

```python
x = tiobe_df['Month']
y = tiobe_df['Python']

plt.figure(figsize=(10, 6))
plt.plot(x, y)

# Rotate x-axis labels
plt.xticks(rotation=45)

# Reduce the number of x-ticks
plt.gca().xaxis.set_major_locator(mdates.MonthLocator(interval=12))  # Show every 12th month, adjust as needed
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))    # Format date labels as 'Year-Month'

plt.xlabel('Month')
plt.ylabel('Python')
# plt.tight_layout()  # Adjust layout to prevent label cut-off
plt.show()
```
