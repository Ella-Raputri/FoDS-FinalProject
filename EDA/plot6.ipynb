```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/merge.csv')
df
```

```python
#Heatmap
correlation_matrix = df.select_dtypes(include=['float64', 'int64']).corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='YlGnBu', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of Numerical Variables")
plt.show()
```

```python
# exclude -1
columns_to_check = ['tiobe index ratings', 'job amount', 'average salary', 'linkedin skill', 
                    'wiki view Oct 2024', 'github user count', 'search count Oct 2024', 
                    'stack_overflow_count_2024']

df_filtered = df[~df[columns_to_check].isin([-1]).any(axis=1)]
df_filtered.info()
```

```python
correlation_matrix = df_filtered.select_dtypes(include=['float64', 'int64']).corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='YlGnBu', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of Numerical Variables (Excluding -1)")
plt.show()
```
