```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/merge.csv')
df
```

```python
# exclude -1
columns_to_check = ['tiobe index ratings', 'job amount', 'average salary', 'linkedin skill', 
                    'wiki view Oct 2024', 'github user count', 'search count Oct 2024', 
                    'stack_overflow_count_2024']

df_filtered = df[~df[columns_to_check].isin([-1]).any(axis=1)]
df_filtered
```

We will plot scatter plot for relations that have correlation coefficient greater than 0.7 and less than -0.7.
##### **1. LinkedIn skill and job amount (correlation coefficient = 0.91)**

```python
plt.figure(figsize=(6,4))

sns.scatterplot(x = 'linkedin skill', y = 'job amount', data = df_filtered, color = 'royalblue', alpha = 0.7)
plt.title("Plot of Job Amount vs LinkedIn Skill", fontsize = 15)
plt.show()
```

##### **2. GitHub user count and job amount (correlation coefficient = 0.81)**

```python
plt.figure(figsize=(6,4))

sns.scatterplot(x = 'github user count', y = 'job amount', data = df_filtered, color = 'royalblue', alpha = 0.7)
plt.title("Plot of GitHub User Count vs Job Amount", fontsize = 15)
plt.show()
```

##### **3. GitHub user count and LinkedIn skill (correlation coefficient = 0.77)**

```python
plt.figure(figsize=(6,4))

sns.scatterplot(x = 'github user count', y = 'linkedin skill', data = df_filtered, color = 'royalblue', alpha = 0.7)
plt.title("Plot of GitHub User Count vs LinkedIn skill", fontsize = 15)
plt.show()
```

##### **4. Stack Overflow user count and job amount (correlation coefficient = 0.89)**

```python
plt.figure(figsize=(6,4))

sns.scatterplot(x = 'stack_overflow_count_2024', y = 'job amount', data = df_filtered, color = 'royalblue', alpha = 0.7)
plt.title("Plot of Stack Overflow User Count vs Job Amount", fontsize = 15)
plt.show()
```

##### **5. Stack Overflow user count and LinkedIn skill (correlation coefficient = 0.85)**

```python
plt.figure(figsize=(6,4))

sns.scatterplot(x = 'stack_overflow_count_2024', y = 'linkedin skill', data = df_filtered, color = 'royalblue', alpha = 0.7)
plt.title("Plot of Stack Overflow User Count vs LinkedIn skill", fontsize = 15)
plt.show()
```

##### **6. Stack Overflow user count and GitHub user count (correlation coefficient = 0.75)**

```python
plt.figure(figsize=(6,4))

sns.scatterplot(x = 'stack_overflow_count_2024', y = 'github user count', data = df_filtered, color = 'royalblue', alpha = 0.7)
plt.title("Plot of Stack Overflow User Count vs GitHub user count", fontsize = 15)
plt.show()
```

### **All Scatter Plots**

```python
fig, axes = plt.subplots(2, 3, figsize=(12,7))

sns.scatterplot(x = 'linkedin skill', y = 'job amount', data = df_filtered, color = 'royalblue', ax=axes[0,0])
axes[0,0].set_title("LinkedIn Skill vs Job Amount")
axes[0,0].set_xlabel("LinkedIn Skill")
axes[0,0].set_ylabel("Job Amount")

sns.scatterplot(x = 'github user count', y = 'job amount', data = df_filtered, color = 'royalblue', ax=axes[0,1])
axes[0,1].set_title("GitHub User Count vs Job Amount")
axes[0,1].set_xlabel("GitHub User Count")
axes[0,1].set_ylabel("Job Amount")

sns.scatterplot(x = 'github user count', y = 'linkedin skill', data = df_filtered, color = 'royalblue', ax=axes[0,2])
axes[0,2].set_title("GitHub User Count vs LinkedIn skill")
axes[0,2].set_xlabel("GitHub User Count")
axes[0,2].set_ylabel("LinkedIn skill")

sns.scatterplot(x = 'stack_overflow_count_2024', y = 'job amount', data = df_filtered, color = 'royalblue', ax=axes[1,0])
axes[1,0].set_title("Stack Overflow User Count vs Job Amount")
axes[1,0].set_xlabel("Stack Overflow User Count")
axes[1,0].set_ylabel("Job Amount")

sns.scatterplot(x = 'stack_overflow_count_2024', y = 'linkedin skill', data = df_filtered, color = 'royalblue', ax=axes[1,1])
axes[1,1].set_title("Stack Overflow User Count vs LinkedIn skill")
axes[1,1].set_xlabel("Stack Overflow User Count")
axes[1,1].set_ylabel("LinkedIn skill")

sns.scatterplot(x = 'stack_overflow_count_2024', y = 'github user count', data = df_filtered, color = 'royalblue', ax=axes[1,2])
axes[1,2].set_title("Stack Overflow User Count vs GitHub user count")
axes[1,2].set_xlabel("Stack Overflow user count")
axes[1,2].set_ylabel("GitHub user count")


plt.tight_layout()
plt.show()
```
