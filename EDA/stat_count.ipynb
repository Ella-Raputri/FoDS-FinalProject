```python
import pandas as pd
```

```python
df = pd.read_csv('../data/merge.csv')
df[df['linkedin skill'] == 0]
```

```python
df.describe()
```

```python
df['job amount'].sum()
```

```python
from sklearn.cluster import KMeans

# Reshape for K-means
kmeans = KMeans(n_clusters=2)
df['job_category'] = kmeans.fit_predict(df[['job amount']])
df['job_category'] = df['job_category'].map({0: 'Low', 1: 'High'})  # Adjust if needed
```

```python
low_and_no_skill = df[(df['job_category'] == 'Low') & (df['linkedin skill'] == 0)]
count = low_and_no_skill.shape[0]
print("Low and 0: ", count)

high_and_no_skill = df[(df['job_category'] == 'High') & (df['linkedin skill'] == 0)]
count = high_and_no_skill.shape[0]
print("High and 0: ",count)

low_and_skill = df[(df['job_category'] == 'Low') & (df['linkedin skill'] != 0)]
count = low_and_skill.shape[0]
print("Low and Non0: ",count)

high_and_skill = df[(df['job_category'] == 'High') & (df['linkedin skill'] != 0)]
count = high_and_skill.shape[0]
print("High and Non0: ", count)
```

```python
low_and_skill = df[(df['job_category'] == 'Low') & (df['github user count'] == -1)]
count = low_and_skill.shape[0]
print("Low and -1: ",count)

high_and_skill = df[(df['job_category'] == 'High') & (df['github user count'] == -1)]
count = high_and_skill.shape[0]
print("High and -1: ", count)

low_and_no_skill = df[(df['job_category'] == 'Low') & (df['github user count'] == 0)]
count = low_and_no_skill.shape[0]
print("Low and 0: ", count)

high_and_no_skill = df[(df['job_category'] == 'High') & (df['github user count'] == 0)]
count = high_and_no_skill.shape[0]
print("High and 0: ",count)

low_and_skill = df[(df['job_category'] == 'Low') & ((df['github user count'] != 0) & (df['github user count'] != -1))]
count = low_and_skill.shape[0]
print("Low and Non0: ",count)

high_and_skill = df[(df['job_category'] == 'High') & ((df['github user count'] != 0) & (df['github user count'] != -1))]
count = high_and_skill.shape[0]
print("High and Non0: ", count)
```
