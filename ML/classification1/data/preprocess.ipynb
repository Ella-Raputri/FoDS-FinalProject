```python
import pandas as pd
```

```python
tiobe_df = pd.read_csv('old/tiobe.csv')
wiki_df = pd.read_csv('old/wiki.csv')
gtrend_df = pd.read_csv('old/gtrend.csv')

languages = ['Java', 'JavaScript', 'PHP', 'Python', 'MySQL', 'SQL', 'C++', 'C', 'PostgreSQL']
```

```python
tiobe_new_df = tiobe_df[['Month','Java', 'JavaScript', 'PHP', 'Python', 'SQL', 'C++', 'C']]
tiobe_new_df
```

```python
wiki_new_df = wiki_df[['Month','Java', 'JavaScript', 'PHP', 'Python', 'MySQL', 'SQL', 'C++', 'C', 'PostgreSQL']]
wiki_new_df
```

```python
gtrend_new_df = gtrend_df[['Month','Java', 'JavaScript', 'PHP', 'Python', 'MySQL', 'SQL', 'C++', 'C', 'PostgreSQL']]
gtrend_new_df
```

```python
datas = [tiobe_new_df, wiki_new_df, gtrend_new_df]
```

```python
results = []

for data in datas: 
    columns_to_normalize = data.columns[1:]

    data_normalized = data.copy()
    data_normalized[columns_to_normalize] = (
        data[columns_to_normalize] - data[columns_to_normalize].min()
    ) / (data[columns_to_normalize].max() - data[columns_to_normalize].min())

    results.append(data_normalized)
```

```python
datas[0].to_csv('new/tiobe.csv', index=False)
datas[1].to_csv('new/wiki.csv', index=False)
datas[2].to_csv('new/gtrend.csv', index=False)
# results[3].to_csv('new/stackoverflow.csv', index=False)
```
