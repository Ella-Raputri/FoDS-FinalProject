```python
import pandas as pd
```

```python
tiobe_df = pd.read_csv('../../data/tiobeindex_complete.csv')
wiki_df = pd.read_csv('../../data/wiki_complete.csv')
gtrend_df = pd.read_csv('../../data/googletrends_complete.csv')
stack_df = pd.read_csv('../../data/stackOverflow_complete.csv')

languages = ['Java', 'JavaScript', 'PHP', 'Python', 'MySQL', 'SQL', 'C++', 'C', 'Visual Basic']
```

```python
tiobe_new_df = tiobe_df[['Month','Java', 'JavaScript', 'PHP', 'Python', 'SQL', 'C++', 'C', 'Visual Basic']]
tiobe_new_df
```

```python
wiki_new_df = wiki_df[['Month','Java', 'JavaScript', 'PHP', 'Python', 'MySQL', 'SQL', 'C++', 'C', 'Visual Basic']]
wiki_new_df
```

```python
gtrend_new_df = gtrend_df[['Month','Java', 'JavaScript', 'PHP', 'Python', 'MySQL', 'SQL', 'C++', 'C', 'Visual Basic']]
gtrend_new_df
```

```python
stack_new_df = stack_df[['year','Java', 'JavaScript', 'PHP', 'Python', 'MySQL', 'SQL', 'C++', 'C', 'Visual Basic']]
stack_new_df
```

```python
datas = [tiobe_new_df, wiki_new_df, gtrend_new_df, stack_new_df]
```

```python
results = []

for data in datas: 
    columns_to_normalize = data.columns[1:]

    data_normalized = data.copy()
    data_normalized[columns_to_normalize] = (
        data[columns_to_normalize] - data[columns_to_normalize].min()
    ) / (data[columns_to_normalize].max() - data[columns_to_normalize].min())

    results.append(data_normalized)
```

```python
results[0].to_csv('new/tiobe.csv', index=False)
results[1].to_csv('new/wiki.csv', index=False)
results[2].to_csv('new/gtrend.csv', index=False)
results[3].to_csv('new/stackoverflow.csv', index=False)
```
