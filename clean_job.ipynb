```python
import pandas as pd
import numpy as np
import re
from sklearn.impute import SimpleImputer
```

```python
df1 = pd.read_csv('jobdata1.csv')
df2 = pd.read_csv('jobdata2.csv')
df3 = pd.read_csv('jobdata3.csv')
df4 = pd.read_csv('jobdata4.csv')
df5 = pd.read_csv('jobdata5.csv')
df6 = pd.read_csv('jobdata6.csv')
df7 = pd.read_csv('jobdata7.csv')
df8 = pd.read_csv('jobdata8.csv')
df9 = pd.read_csv('jobdata9.csv')
df10 = pd.read_csv('jobdata10.csv')
df11 = pd.read_csv('jobdata11.csv')
df12 = pd.read_csv('jobdata12.csv')
df13 = pd.read_csv('jobdata13.csv')
df = pd.concat([df1,df2,df3,df4,df5,df6,df7,df8,df9,df10,df11,df12,df13], sort=False)
df.rename(str.lower, axis='columns',inplace=True)
df.describe()
```

```python
df.drop_duplicates(inplace=True)
df = df.dropna(subset=['job title'])
df.info()
```

```python
def cleanTitle(txt):
    x = re.sub(r"\s*-\s*job post$", "", txt, flags=re.IGNORECASE).strip().lower()
    return x

df['job title'] = df['job title'].apply(cleanTitle)

df['job title']
```

```python
language_df = pd.read_csv('nama.csv')
languages = language_df['title'].tolist()

def extract_languages(job_description):
    found_languages = []
    for language in languages:
        if re.search(fr'\b{language}\b', job_description, re.IGNORECASE):
            found_languages.append(language)
    return found_languages

df['programming language'] = df['job description'].apply(extract_languages)
df.head()
```

```python
df = df[df['programming language'].str.len() > 0].reset_index(drop=True)
df.head()
```

```python
def cleanSalary(txt):
    if isinstance(txt, str):
        numbers = re.findall(r'\d{1,3}(?:,\d{3})*', txt)
        if numbers:
            max_salary = int(numbers[-1].replace(',', ''))
            return max_salary
        return None
    return txt

df['salary'] = df['salary'].apply(cleanSalary)
df.head()
```

```python
df.info()
```

```python
num_cols = df.select_dtypes(include=['float64']).columns
imp = SimpleImputer(strategy='mean')
df[num_cols] = imp.fit_transform(df[num_cols])

df.head()
```

```python
remove_list = [
    'T', '.QL', 'min', 'D', 'Flow', 'Clean', 'E', 'world', 'USD', 'CHAIN',
    'D++', 'Plus', 'FOCUS', 'SIMPLE', 'never', 'video', 'Link', 'Address',
    'Self', 'S3', 'Ada', 'GAME', 'Icon', 'Low*', 'H++', 'ML', 'Expect',
    'Copilot', 'potential', 'pipelines', 'solid', 'Processing',
    'RAPID', 'lain', 'april', 'Links', 'links', 'Fun', 'Nice', 'flagship',
    'Small', 'Lean', 'Simplicity', 'Scratch', 'R++', 'Pizza', 'F', 'M', 'lift',
    'Ease', 'Revolution', 'MATLAB', 'Click', 'HOPE', 'Apex', 'Adept', 'I',
    'ten', 'dss', 'Pure', 'GAP', 'dio', 'Falcon', 'Alpha', 'Plot', 'Joy',
    'CONVERT', 'Frank', 'P', 'lever', 'JADE', 'Blue', 'Inform', 'NATURAL',
    'U', 'Reason', 'ARM', 'SR', 'Crystal', 'PowerHouse', 'Conceptual', 'Factor',
    'hello', 'l', 'False', 'tab', 'RPL', 'slim', 'cx', 'fe', 'Chrome', 'none',
    'Tao', 'Flavors', 'Cyber', 'Ad-hoc', 'eC', 'Io', 'Bash', 'DEMOS', 'Q',
    'Pro*C', 'Gist', 'DIAGRAM', 'PILOT', 'REDUCE', 'Monkey', 'little', 'Forth',
    'Z++', 'Z', 'abs', 'liquidity', 'FaCT', 'Bon', 'MACRO', 'FLUX', 'Argus',
    'V', 'SCM', 'Red', 'CHILL', 'MetaL', 'Puppet', 'SaltStack', 'bamboo',
    'Mercury', 'SIGNAL', 'Spatial', 'Pig Latin', 'AL', 'PowerBuilder', 'SRL',
    'Logos', 'Dale', 'Virgil', 'scopes', 'LUCID', 'CUDA', 'LINQ', 'oil', 'JOVIAL',
    'APL', 'Cell', 'rocket', 'MIMIC', 'Charity', 'Meson', 'DYNAMO', 'PLANNER',
    'Gherkin', 'frame','F*','P*','A++','A+','NPL','kernel','K','S','JS++','D4','OBJ',
    'Delphi','FoxPRO','Visual FoxPro','Scala','Objective C++','B','SAS','SPSS',
    'Scheme','jasper','Amazon Redshift','PROMETHEUS','PL/SQL'
]

def remove_unwanted_languages(lang_list):
    return [lang for lang in lang_list if lang not in remove_list]

df['programming language'] = df['programming language'].apply(remove_unwanted_languages)
```

```python
def basic_combine(lang_list):
    return ['Visual Basic' if lang in ['BASIC Programming', 'BASIC'] else lang for lang in lang_list]

def js_typofix(lang_list):
    return ['JavaScript' if lang in ['JS'] else lang for lang in lang_list]

def python_typofix(lang_list):
    return ['Python' if lang in ['Phyton'] else lang for lang in lang_list]

def remove_duplicates(lang_list):
    return list(set(lang_list))

df['programming language'] = df['programming language'].apply(basic_combine)
df['programming language'] = df['programming language'].apply(js_typofix)
df['programming language'] = df['programming language'].apply(python_typofix)
df['programming language'] = df['programming language'].apply(remove_duplicates)
```

```python
d = {}

def note_numlanguage(languages):
    for lang_list in languages:
        for lang in lang_list:  
            if lang in d:
                d[lang] += 1  
            else:
                d[lang] = 1  

note_numlanguage(df['programming language'])
print(len(d))
for key, value in d.items():
    print(f"{key}: {value}")
```
