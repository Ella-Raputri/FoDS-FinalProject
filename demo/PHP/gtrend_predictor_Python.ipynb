```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import MinMaxScaler
from imblearn.over_sampling import SMOTE
```

```python
language = 'PHP'
df = pd.read_csv(f'../../data/normalized/{language}.csv')

df[f'{language}_gtrend_label'] = df[f'{language}_gtrend_label'].astype('category')
df[f'{language}_gtrend_label_encoded'] = df[f'{language}_gtrend_label'].cat.codes

X = df[[f'{language}_wiki', f'{language}_tiobe']]
y = df[f'{language}_gtrend_label_encoded']

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

X_train1, X_test, y_train1, y_test = train_test_split(X_scaled, y, test_size=0.5, random_state=42, stratify=y)
smote = SMOTE(random_state=42)
X_train, y_train = smote.fit_resample(X_train1, y_train1)

models = {
    "Random Forest": RandomForestClassifier(bootstrap=False, max_depth=None, min_samples_leaf=2, min_samples_split=2, n_estimators=300),
    "Decision Tree": DecisionTreeClassifier(criterion='gini', max_depth=None, min_samples_leaf=1, min_samples_split=10, splitter='best'),
    "XGBoost": XGBClassifier(colsample_bytree=0.8, learning_rate=0.01, max_depth=3, min_child_weight=2, n_estimators=100, subsample=1.0),
    "LightGBM": LGBMClassifier(colsample_bytree=0.8, learning_rate=0.01, max_depth=3, min_child_samples=20, n_estimators=200, subsample=0.8),
    "SVC": SVC(C=10, kernel='rbf',probability=True),
    "KNN": KNeighborsClassifier(n_neighbors=3),
    "MLP": MLPClassifier(activation='relu', hidden_layer_sizes=(100,), max_iter=500),
    "Naive Bayes": GaussianNB(),
    "Logistic Regression": LogisticRegression(C=0.1)
}


for model_name, model in models.items():
    model.fit(X_train, y_train)
```

```python
def predict_user_input(models, language, df):
    print("Enter the following feature values (normalized):")
    wiki = float(input(f"Enter the {language}_wiki value (0 to 1): "))
    tiobe = float(input(f"Enter the {language}_tiobe value (0 to 1): "))
    
    user_input = pd.DataFrame([[wiki, tiobe]], columns=[f'{language}_wiki', f'{language}_tiobe']) 
    
    predictions = {}
    for model_name, model in models.items():
        pred_encoded = model.predict(user_input)[0]  
        pred_label = df[f'{language}_gtrend_label'].cat.categories[pred_encoded]  
        predictions[model_name] = pred_label
    
    for model_name, prediction in predictions.items():
        print(f"{model_name} prediction: {prediction}")


predict_user_input(models, language, df)
```
