```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('googletrends2.csv')
df
```

```python
df['Month'] = pd.to_datetime(df['Month'])
df_2024 = df[(df['Month'].dt.year == 2024) & (df['Month'].dt.month == 10)]
df_2024
```

```python
df_melted = df_2024.drop(columns=['Month']).melt(var_name='programming language', value_name='search count Oct 2024')
language_totals = df_melted.groupby('programming language')['search count Oct 2024'].sum().reset_index()
language_totals
```

```python
language_totals.to_csv('gtrends_10_2024.csv',index=False)
```
