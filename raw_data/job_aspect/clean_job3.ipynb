```python
import pandas as pd
import re
import ast
from sklearn.impute import SimpleImputer
```

```python
df1=pd.read_csv('language_list.csv')
df2 = pd.read_csv('language_list_2.csv')
df = pd.merge(df1,df2[['programming language','location','job title']],on='programming language', how='left')
df.drop(columns='location_x', inplace=True)
df.rename(columns ={'location_y': 'location'}, inplace=True)

df.head()
```

```python
job_title_need = ['manager','full stack', 'front end','frontend','backend', 'back end', 'software','mobile','devops','game','QA','quality assurance','data scientist', 'data analyst','machine learning','ML','artificial intelligence','AI','programmer','analyst','developer','engineer']

for job in job_title_need:
    df[job] =0
df['others'] =0

df.head()
```

```python
df['job title'] = df['job title'].apply(ast.literal_eval)
```

```python
def categorize_titles_in_row(row):
    title_list = row['job title']  
    counts = {job: 0 for job in job_title_need}  
    counts['others'] = 0 

    for title in title_list:
        title_lower = title.lower()
        matched = False
        for job in job_title_need:
            if job in title_lower:
                counts[job] += 1
                matched = True
                break
        if not matched:
            counts['others'] += 1

    for job in counts:
        row[job] = counts[job]
    return row

df = df.apply(categorize_titles_in_row, axis=1)

df.head()
```

```python
df['front end'] = df['front end'] + df['frontend']
df['back end'] = df['back end'] + df['backend']
df['QA'] = df['QA'] + df['quality assurance']
df['AI'] =df['artificial intelligence']+df['AI']
df['machine learning'] =df['machine learning']+df['ML']

df = df.drop(columns=['job title', 'frontend', 'backend', 'quality assurance','artificial intelligence','ML'])
df.head()
```

```python
df.rename(columns = {'software':'software developer/engineer', 'mobile':'mobile developer','game':'game developer',
                     'analyst':'other analyst', 'developer':'other developer','AI':'AI engineer',
                     'QA':'QA engineer','machine learning':'machine learning engineer','engineer':'other engineer'}, inplace = True)

df.head()
```

```python
df.to_csv('job_result2.csv', index=False)
```
