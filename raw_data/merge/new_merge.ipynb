```python
import pandas as pd
```

```python
df = pd.read_csv('result3.csv')
df[df['programming language'] == 'Python']
```

```python
df.drop(columns=['avg wiki views (monthly)', 'average search count','stack overflow count'], inplace=True)
df
```

```python
wiki_df = pd.read_csv('../wiki/wikicount_10_2024.csv')
stack_df = pd.read_csv('../stack_overflow/yearly_stack_overflow_count.csv')
gtrend_df = pd.read_csv('../google_trends/gtrends_10_2024.csv')
```

```python
stack_df = stack_df.filter(items=['programming language','stack_overflow_count_2024'])
stack_df
```

```python
res1 = pd.merge(df, wiki_df, on="programming language",  how='left', suffixes=('_df', '_wiki'))
res1.fillna(-1, inplace=True)
res1.head()
```

```python
res2 = pd.merge(res1, gtrend_df, on="programming language")
res3 = pd.merge(res2, stack_df, on="programming language")
```

```python
res3.to_csv('result4.csv', index=False)
```
