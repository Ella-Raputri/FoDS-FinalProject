```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('cleaned_wikidata22.csv')
df
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df_2024 = df[(df['Date'].dt.year == 2024) & (df['Date'].dt.month == 10)]
df_2024
```

```python
df_melted = df_2024.drop(columns=['Date']).melt(var_name='programming language', value_name='wiki view Oct 2024')
language_totals = df_melted.groupby('programming language')['wiki view Oct 2024'].sum().reset_index()
language_totals
```

```python
language_totals.to_csv('wikicount_10_2024.csv',index=False)
```
