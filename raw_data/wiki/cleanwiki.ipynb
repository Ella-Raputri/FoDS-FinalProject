```python
import pandas as pd
import os
```

```python
folder_path = 'wikidata'
dataframes = []
```

```python
dataframes = []

for file in os.listdir(folder_path):
    file_path = os.path.join(folder_path, file)  
    try:
        df = pd.read_csv(file_path)  
        if not df.empty:  
            dataframes.append(df) 
        else:
            print(f"Warning: The file {file} is empty.")
    
    except Exception as e:
        print(f"Error reading {file}: {e}")
```

```python
if dataframes:
    combined_df = dataframes[0]

    for df in dataframes[1:]:
        combined_df = combined_df.merge(df, on='Date', how='outer')    
    
else:
    print("No valid CSV files were found to merge.")
```

```python
combined_df.info()
```

```python
combined_df.columns = combined_df.columns.str.replace(r"\s*\(.*\)", "", regex=True)
combined_df.info()
```

```python
combined_df.rename(columns={"C Sharp":"C#"}, inplace=True)
combined_df.rename(columns={"Bahasa rakitan":"Assembly"}, inplace=True)
combined_df.rename(columns={"Embarcadero Delphi":"Delphi"}, inplace=True)
combined_df.rename(columns={"Visual Basic for Applications":"VBA"}, inplace=True)
combined_df.rename(columns={"Statistical Analysis System":"SAS"}, inplace=True)

combined_df.info()
```

```python
combined_df.to_csv('cleaned_wikidata22.csv', index=False)
```

```python
averages = []

for col in combined_df.columns[1:]:  
    first_non_zero_idx = combined_df[col].ne(0).idxmax()
    
    if first_non_zero_idx is not None:
        avg_views = combined_df.loc[first_non_zero_idx:, col].mean()
        averages.append({"name": col, "avg wiki views (monthly)": avg_views})
```

```python
avg_df = pd.DataFrame(averages)
avg_df.to_csv('avg_wiki_monthly22.csv',index=False)
avg_df
```
